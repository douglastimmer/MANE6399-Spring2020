```python
# Python cell to read data from file into dataframe df
import pandas as pd
df=pd.read_csv('column_2C.DAT',delimiter=' ')
df.columns=['pIncidence','pTilt','llAngle','sSlope','pRadius','spon','classification']
#print(df)
```

```python
# Problem 1 
# Enter the Python code to solve problem 1 in this cell
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import graphviz 
from sklearn import tree
from IPython.display import display, HTML, Latex
display(HTML("<h1>Problem 1</h1>"))
df1=df.dropna(axis=0,how='any')       # remove rows with missing values
# create endogenous and exogenous variables
X= np.array(df1.iloc[:, 0:6])
#print(X)
y = np.array(df1['classification'])
#print(y)
# split and transform data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
#
clf1 = DecisionTreeClassifier()
clf1.fit(X_train,y_train)
display(HTML("<h2>Accuracy Results</h2>"))
print("training set accuracy is %f"%clf1.score(X_train,y_train))
print("testing set accuracy is %f"%clf1.score(X_test,y_test))
y_pred_train=clf1.predict(X_train)
display(HTML("<h2>Confusion Matrices</h2>"))
cm=confusion_matrix(y_train,y_pred_train)
print("confusion matrix for training set")
print(cm)
y_pred_test=clf1.predict(X_test)
cm=confusion_matrix(y_test,y_pred_test)
print("confusion matrix for test set")
print(cm)
display(HTML("<h2>Graphviz Representation of Decision Tree</h2>"))
dot_data = tree.export_graphviz(clf1, out_file=None) 
graph = graphviz.Source(dot_data)  
graph
```

```python
# Problem 2
# Enter the Python code to solve problem 2 in this cell
display(HTML("<h1>Problem 2</h1>"))
display(HTML("<p><b>Reason:</b> The parameter selected is max_depth for the solutions. The intention is to simplify the tree</p>"))
#
clf2 = DecisionTreeClassifier(max_depth=4)
clf2.fit(X_train,y_train)
display(HTML("<h2>Accuracy Results</h2>"))
print("training set accuracy is %f"%clf2.score(X_train,y_train))
print("testing set accuracy is %f"%clf2.score(X_test,y_test))
y_pred_train=clf2.predict(X_train)
display(HTML("<h2>Confusion Matrices</h2>"))
cm=confusion_matrix(y_train,y_pred_train)
print("confusion matrix for training set")
print(cm)
y_pred_test=clf2.predict(X_test)
cm=confusion_matrix(y_test,y_pred_test)
print("confusion matrix for test set")
print(cm)
display(HTML("<h2>Graphviz Representation of Decision Tree</h2>"))
dot_data = tree.export_graphviz(clf2, out_file=None) 
graph = graphviz.Source(dot_data)  
graph
display(HTML("<p><b>Impact:</b> The decision tree model is greatly simplified. The performance, as expected, was not as good as the model in problem 1. However, it is interesting to note that the test set accuracy was actually improved in the reduced model. This is an indication of overfitting in the first model.</p>"))
display(HTML("<p><b>NOTE: </b>This is not the only parameter that could have been selected.</p>"))
```

```python
# Problem 3
# Enter the Python code to solve problem 3 in this cell
display(HTML("<h1>Problem 3</h1>"))
from sklearn.ensemble import GradientBoostingClassifier
clf3 = GradientBoostingClassifier()
clf3.fit(X_train,y_train)
print("training set accuracy is %f"%clf3.score(X_train,y_train))
print("testing set accuracy is %f"%clf3.score(X_test,y_test))
display(HTML("<h2>Accuracy Results</h2>"))
y_pred_train=clf3.predict(X_train)
display(HTML("<h2>Confusion Matrices</h2>"))
cm=confusion_matrix(y_train,y_pred_train)
print("confusion matrix for training set")
print(cm)
y_pred_test=clf3.predict(X_test)
cm=confusion_matrix(y_test,y_pred_test)
print("confusion matrix for test set")
print(cm)
```

```python
# Problem 4
# Enter the Python code to solve problem 4 in this cell
display(HTML("<h1>Problem 4</h1>"))
display(HTML("<p><b>Parameter and Reason: </b> The paramter min_samples_leaf was set to 5 instead of the default value of 1. The goal is to simplify the tree without sacrificing accuracy.</p>"))
from sklearn.ensemble import GradientBoostingClassifier
clf4 = GradientBoostingClassifier(min_samples_leaf=5)
clf4.fit(X_train,y_train)
print("training set accuracy is %f"%clf4.score(X_train,y_train))
print("testing set accuracy is %f"%clf4.score(X_test,y_test))
display(HTML("<h2>Accuracy Results</h2>"))
y_pred_train=clf4.predict(X_train)
display(HTML("<h2>Confusion Matrices</h2>"))
cm=confusion_matrix(y_train,y_pred_train)
print("confusion matrix for training set")
print(cm)
y_pred_test=clf4.predict(X_test)
cm=confusion_matrix(y_test,y_pred_test)
print("confusion matrix for test set")
print(cm)
display(HTML("<p><b>Impact:</b>The accuracy performance for the test set and the confusion matrix for the test set are only slightly worse; so there is not a big drop in performance. Without graphing the trees (not part of the assignment), it is hard to determine if the model was simplified </p>"))
```

```python
# Problem 6
# Enter the Python code to solve problem 6 in this cell
display(HTML("<h1>Problem 6</h1>"))
display(HTML("<p><b>Parameter and Reason: </b> The paramter min_samples_split was set to 5 instead of the default value of 1. The goal is to simplify the tree without sacrificing accuracy.</p>"))

clf6 = RandomForestClassifier(min_samples_split=5)
clf6.fit(X_train,y_train)
display(HTML("<h2>Accuracy Results</h2>"))
print("training set accuracy is %f"%clf6.score(X_train,y_train))
print("testing set accuracy is %f"%clf6.score(X_test,y_test))
display(HTML("<h2>Confusion Matrices</h2>"))
y_pred_train=clf6.predict(X_train)
cm=confusion_matrix(y_train,y_pred_train)
print("confusion matrix for training set")
print(cm)
y_pred_test=clf6.predict(X_test)
cm=confusion_matrix(y_test,y_pred_test)
print("confusion matrix for test set")
print(cm)
display(HTML("<p><b>Impact:</b>The accuracy performance for the test set and the confusion matrix for the test set are only slightly worse; so there is not a big drop in performance. Without graphing the trees (not part of the assignment), it is hard to determine if the model was simplified </p>"))
```
