```python
# Cell Two - Part A
import pandas as pd
from IPython.display import HTML, display
import matplotlib.pyplot as plt

df = pd.read_excel(open('wdbc_data.xlsx','rb'))
display(HTML("<h1>Cell Two</h1>"))
display(HTML("<h2>Descriptive Statistics"))
print(df.describe())
display(HTML("<h2>Scatter Matrix</h2>"))
pd.plotting.scatter_matrix(df,alpha=0.2, figsize=(10,10),diagonal='hist')
plt.show()
display(HTML("<h2>Correlation Matrix</h2>"))
print(df.corr())
```

# Cell Three - Part B
## Multicollinearyity Question

Multicollinearity is a large problem with this data set that is seen in the scatter matrix and correlation analysis. In the correlation plot, there are many pairs of variables that plot almost as a straight line. If a correlation of 0.8 is used as a cutoff, there are nine pairs of variables that are highly correlated.

```python
# Cell Four - Part C
from sklearn import preprocessing
display(HTML("<h1>Cell Four</h1>"))
display(HTML("<h2>Standardization</h2>"))
std_scaler=preprocessing.StandardScaler().fit(df.values)
df_scaled=std_scaler.transform(df)
df_scaled_df=pd.DataFrame(df_scaled,index=df.index,columns=df.columns)
df_scaled_df.describe()
```

# Cell Five - Part D

Notice that the mean of all variables is essentially zero and the standard deviation (std) is 1.0.

```python
# Cell Six - Part E
import numpy as np
display(HTML("<h1>Cell Six - Part E"))
display(HTML("<h2>PCA</h2>"))
from sklearn import decomposition
#set up and execute PCA
pca=decomposition.PCA(n_components=10)
pca.fit(df_scaled_df)
#display eigen values
display(HTML("<h3>Eigenvalues (sorted)</h3>"))
print(pca.explained_variance_)
#display explained variance ratio
display(HTML("<h3> Percent variability explained by each component</h3>"))
print(pca.explained_variance_ratio_)
#display cumulative sum of variability explained
display(HTML("<h3> Cumulative Percent variability explained by each component</h3>"))
print(np.cumsum(pca.explained_variance_ratio_))
display(HTML("<h3>Scree Plot</h3>"))
plt.plot(np.arange(0,10,1),pca.explained_variance_,'-bo')
plt.show()
```

# Cell Seven - Part F

## Eigen value Criterion

The eigen value criterion is to use any eigen vectors whose eigen value is > 1. For this problem, the eigen value criterion would recommend two eigen vectors.

## Percent Variability Explained

This criterion requires the user to specify a required percent of variability explained. Suppose 95 percent was selected. In this case, the first five eigen vectors are required.

## Scree Plot

This is a completely subjective decision, I would recommend the first four or five eigen vectors be used.

```python
# Cell Eight - Part G
display(HTML("<h1>Cell Eight - Part G</ha>"))
pca2=decomposition.PCA(n_components=5)
pca2.fit(df_scaled_df)
#display eigen values
display(HTML("<h3>Eigenvalues (sorted)</h3>"))
print(pca2.explained_variance_)
#display explained variance ratio
display(HTML("<h3> Percent variability explained by each component</h3>"))
print(pca2.explained_variance_ratio_)
#display cumulative sum of variability explained
display(HTML("<h3> Cumulative Percent variability explained by each component</h3>"))
print(np.cumsum(pca2.explained_variance_ratio_))
display(HTML("<h3>Eigen Vector</h3>"))
print(pca2.components_)
```
