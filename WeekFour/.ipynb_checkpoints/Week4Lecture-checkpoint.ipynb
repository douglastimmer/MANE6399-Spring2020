```python
# Cell 2
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
yachtData = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/00243/yacht_hydrodynamics.data",
                   delim_whitespace = True, header=None,
                   names = ['cb', 'prismatic', 'ld_ratio', 'bd_ratio', 'lb_ratio', 'froude'],na_values=['?'])
fig = plt.figure()
#first subplot
ax=fig.add_subplot(2,3,1)
scipy.stats.probplot(yachtData['cb'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('cb')
#second subplot
ax=fig.add_subplot(2,3,2)
scipy.stats.probplot(yachtData['prismatic'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('prismatic')
#third subplot
ax=fig.add_subplot(2,3,3)
scipy.stats.probplot(yachtData['ld_ratio'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('ld_ratio')
#fourth subplot
ax=fig.add_subplot(2,3,4)
scipy.stats.probplot(yachtData['bd_ratio'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('bd_ratio')
#fifth subplot
ax=fig.add_subplot(2,3,5)
scipy.stats.probplot(yachtData['lb_ratio'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('lb_ratio')
#sixth subplot
ax=fig.add_subplot(2,3,6)
scipy.stats.probplot(yachtData['froude'],dist=scipy.stats.norm,plot=plt)
ax.set_xlabel('froude')
plt.show()
```

```python
# Cell 3
count=1
fig2 = plt.figure()
for c in yachtData.columns:
    ax=fig2.add_subplot(2,3,count)
    scipy.stats.probplot(yachtData[c],dist=scipy.stats.norm,plot=plt)
    ax.set_xlabel(c)
    count=count+1
plt.show()
```

```python
# Cell 4
count=1
fig3 = plt.figure()
fig3.subplots_adjust(hspace=0.8, wspace=0.8)
for c in yachtData.columns:
    ax=fig3.add_subplot(2,3,count)
    scipy.stats.probplot(yachtData[c],dist=scipy.stats.norm,plot=plt)
    ax.set_xlabel(c)
    count=count+1
plt.show()
```

```python
# Cell 5
import pandas as pd
df = pd.read_excel('Concrete_Data.xls', sheetname='Sheet1')
df.describe()
df.corr()
```

```python
# Cell 6
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
scatter_matrix(df, diagonal='kde')
plt.show()
```

```python
# Cell 7
from IPython.display import display, Math, Latex, HTML
import statsmodels.formula.api as sm
display(HTML("<h1> Simple Linear Regression using Statmodels"))
#set up variables holding the x and y variables

model1 = sm.ols(formula='strength~cement',data=df)
fitted1 = model1.fit()
print(fitted1.summary())
```

```python
# Cell 8
display(HTML("<h1> Plot Fitted Model"))
plt.plot(df["cement"], df["strength"], 'ro')
plt.plot(df["cement"], fitted1.fittedvalues, 'b')
plt.legend(['Data', 'Fitted model'])
#plt.ylim(0, 10)
#plt.xlim(-2, 12)
plt.xlabel('cement')
plt.ylabel('strength')
plt.title('Simple Linear Regression')
plt.show()
```

```python
# Cell 9
display(HTML("<h1> Residual Plot 1"))
residual=fitted1.resid
plt.plot(df["cement"], residual, 'ro')
plt.xlabel('cement')
plt.ylabel('residual')
plt.title('Residuals versus cement')
plt.show()
```

```python
# cell 10
display(HTML("<h1> Residual Plot 2"))
residual2=fitted1.resid_pearson
plt.plot(df["cement"], residual2, 'ro')
plt.xlabel('cement')
plt.ylabel('standardized residual')
plt.title('Standardized Residuals versus cement')
plt.show()
```

```python
# cell 11
display(HTML("<h1> Residual Plot 3"))
fittedValues=fitted1.fittedvalues
plt.plot(fittedValues, residual2, 'ro')
plt.xlabel('fitted values')
plt.ylabel('standardized residual')
plt.title('Standardized Residuals versus fitted values')
plt.show()
```

```python
# Cell 12
import numpy as np

id=np.arange(1,df["cement"].size+1,1)
plt.plot(id,residual2,'-bo')
plt.show()
```

```python
# Cell 13
from scipy import stats
# normal probability plot
stats.probplot(residual, plot=plt)
plt.title('Normal Probability Plot')
plt.show()
```

```python
# Cell 14
from statsmodels.stats.outliers_influence import OLSInfluence
test_class = OLSInfluence(fitted1)
plt.stem(np.arange(len(test_class.hat_matrix_diag)),test_class.hat_matrix_diag)
plt.show()
```

```python
# Cell 15
plt.stem(np.arange(len(test_class.cooks_distance[0])),test_class.cooks_distance[0])
plt.show()
```
