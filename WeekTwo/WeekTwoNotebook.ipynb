```python
# Panda - Example 6
import pandas as pd
data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original",
                   delim_whitespace = True, header=None,
                   names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
                            'model', 'origin', 'car_name'],na_values=['?'])
print(data.head(3))
print(data.tail(3))
```

```python
# Panda - Example 7
print(data.describe())
print("")
print("""variance of mpg is %f"""%(data["mpg"].var()))
# discuss difference between issuing commands and printing 
```

```python
# Panda - Example 8
print("Covariance matrix")
print(data.cov())
print("")
print("Correlation matrix")
print(data.corr())
```

```python
# Panda - Example 9
import matplotlib.pyplot as plt
pd.DataFrame.hist(data)
plt.show()
#
```

```python
# Panda - Example 10
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
# Panda - Example 11
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show
```

```python
# Panda - Example 12
import scipy
scipy.stats.probplot(data["mpg"], dist=scipy.stats.norm,plot=plt);
plt.show()
```

```python
# Panda - Example 13
dataTrimmed=data.dropna(axis=0,how='any')
scipy.stats.probplot(dataTrimmed["mpg"],dist=scipy.stats.norm,plot=plt)
plt.show()
```

```python
# Panda - Example 14
pd.plotting.scatter_matrix(dataTrimmed,alpha=0.2, figsize=(6,6),diagonal='kde')
plt.show()
# discuss changing figsize and diagonal
```
